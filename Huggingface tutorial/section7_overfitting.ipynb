```python
from datasets import load_dataset

# Dataset id from huggingface.co/dataset
dataset_id = "burtenshaw/PleIAs_common_corpus_code_classification"

# Load raw dataset
dataset = load_dataset(dataset_id)
```

```python
print(len(dataset["train"]))
print(dataset["train"][0])
```

```python
from transformers import AutoTokenizer

# Model id to load the tokenizer
model_id = "answerdotai/ModernBERT-base"

# Load Tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Tokenize helper function
def tokenize(batch):
    return tokenizer(batch['text'], padding=True, truncation=True, return_tensors="pt")

# Tokenize dataset
tokenized_dataset = dataset.map(tokenize, batched=True, remove_columns=["text"])

tokenized_dataset["train"].features.keys()
# dict_keys(['labels', 'input_ids', 'attention_mask'])
```

```python
from transformers import AutoModelForSequenceClassification

# Model id to load the tokenizer
model_id = "answerdotai/ModernBERT-base"

# Prepare model labels - useful for inference
labels = list(set(tokenized_dataset["train"]["labels"]))
num_labels = len(labels)
label2id, id2label = dict(), dict()
for i, label in enumerate(labels):
    label2id[label] = str(i)
    id2label[str(i)] = label
```

```python
# 1. Collect unique labels from ALL dataset splits, not just 'train'
unique_labels = set()
for split in tokenized_dataset.keys():
    unique_labels.update(tokenized_dataset[split]["labels"])

labels = list(unique_labels)
num_labels = len(labels)

# 2. Rebuild the dictionaries with the complete list
label2id, id2label = dict(), dict()
for i, label in enumerate(labels):
    label2id[label] = str(i)
    id2label[str(i)] = label

# 3. Define the encoding function using the COMPLETE dictionary
def encode_labels(example):
    example["labels"] = int(label2id[example["labels"]])
    return example

# 4. Now apply the mapping
tokenized_dataset = tokenized_dataset.map(encode_labels)

print(f"Success! Mapped {num_labels} unique labels.")
```

```python
# Download the model from huggingface.co/models
untrained_model = AutoModelForSequenceClassification.from_pretrained(
    model_id, num_labels=num_labels, label2id=label2id, id2label=id2label,
)
```

```python
import copy

def get_untrained_model():
  return copy.deepcopy(untrained_model)
```

```python
model = get_untrained_model()
```

```python
import numpy as np
from sklearn.metrics import f1_score
#harmonic mean of Precision and Recall

# Metric helper method
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    score = f1_score(
            labels, predictions, labels=labels, pos_label=1, average="weighted"
        )
    return {"f1": float(score) if score == 1 else score}
```

```python
from huggingface_hub import HfFolder
from transformers import Trainer, TrainingArguments

# Define training args
training_args = TrainingArguments(
    output_dir= "ModernBERT-code-classifier",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=16,
    learning_rate=5e-5,
    num_train_epochs=5,
    bf16=True,                  # T4 does not support bf16 well
    # fp16=True,                   # Use mixed precision fp16 (Standard for T4)
    optim="adamw_torch_fused",   # improved optimizer to be faster on GPU by kernel fusion
    gradient_checkpointing=True, # CRITICAL: Saves massive VRAM by discarding intermediate activations# logging & evaluation strategies
    logging_strategy="steps",
    logging_steps=100,
    eval_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    # push to hub parameters
    push_to_hub=False,
    report_to="none"
)

```

# Overfitting

```python
limited_dataset = tokenized_dataset["train"].select(range(10))

# Create a Trainer instance
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=limited_dataset,
    eval_dataset=tokenized_dataset["test"],
    compute_metrics=compute_metrics,
)
trainer.train()
```

```python
# clear memory

import torch
torch.cuda.empty_cache()

# reset parameters

del trainer
model = get_untrained_model()
limited_dataset = tokenized_dataset["train"].select(range(10))
```
